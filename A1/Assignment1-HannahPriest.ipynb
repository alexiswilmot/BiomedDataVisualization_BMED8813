```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


file_path = "medical_costs.csv"
medical_data = pd.read_csv(file_path)

medical_data.head()
```

```python
# plot 1 - smokers vs. non-smokers:
smoker_proportions = medical_data["Smoker"].value_counts(normalize=True)
plt.figure(figsize=(8,8))
plt.pie(smoker_proportions, labels = smoker_proportions.index, startangle = 90,  autopct='%1.1f%%', colors = ("lightblue", "salmon"))
plt.title("Proportion of Smokers vs. Non-Smokers")
plt.axis("equal")
plt.show()
```

```python
# plot 2 - proportion of smokers vs nonsmokers by region:
proportions = pd.crosstab(medical_data["Region"], medical_data["Smoker"], normalize="index")
ax = proportions.plot(kind = "bar", stacked = True, figsize = (10,6), color = ["lightblue", "salmon",])
sns.set(style = "whitegrid")
plt.title("Proportion of Smokers vs. Non-Smokers by Region")
plt.xlabel("Region")
plt.ylabel("Proportion")
plt.legend(title = "Smoker")
plt.tight_layout()
plt.show()
```

```python
# plot 3 - average medical costs by age:
avg_cost_by_age = medical_data.groupby("Age")["Medical Cost"].mean()
plt.figure(figsize=(10,6))
plt.plot(avg_cost_by_age.index, avg_cost_by_age.values, color = "green", marker = "o")
sns.set(style = "whitegrid")
plt.title("Average Medical Cost by Age")
plt.xlabel("Age")
plt.ylabel("Average Medical Cost")
plt.tight_layout()
plt.show()
```

```python
# Part 2:
# plot 1 - medical costs across ages:
total_cost_by_age = medical_data.groupby("Age")["Medical Cost"].sum()
plt.figure(figsize=(12, 6))
sns.set(style = "whitegrid")
plt.bar(total_cost_by_age.index, total_cost_by_age.values, color = "lightblue")
plt.title("Distribution of Medical Costs Across Ages")
plt.xlabel("Age (years)")
plt.ylabel("Total Medical Costs ($)")
plt.tight_layout()
plt.show()
```

```python
# plot 2 - medical costs by BMI:

sns.set(style = "whitegrid")
# separating numerical vs categorical 
numerical_cols = ["Age", "BMI", "Children"]
categorical_cols = ["Sex", "Smoker", "Region"]

sns.regplot(x=medical_data["BMI"], y=medical_data["Medical Cost"], color = "salmon")
plt.title(f'{"BMI"} vs Medical Cost')

plt.show 
```

```python
# plot 3 - distribution of medical costs by region: 
# create subplots for categorical columns

sns.boxplot(x=medical_data["Region"], y=medical_data["Medical Cost"], color = "skyblue")
plt.title("Distribution of Medical Costs by Region")
plt.xlabel("Region")
plt.ylabel("Medical Costs ($)")
plt.tight_layout(rect = [0, 0, 1, 0.96])
plt.show()
```
