```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('medical_costs.csv')
age = df.Age 
cost = df.MedicalCost
smokeStatus = df.Smoker
df

```

```python
# Pie chart of smoker status
smokerCounts = df['Smoker'].value_counts()

plt.figure()
plt.pie(smokerCounts, labels=smokerCounts.index, autopct='%1.1f%%', startangle=90, colors=['lightskyblue', 'lightcoral'])
plt.title('Proportion of Smokers vs. Non-Smokers')
plt.show()
```

```python
# line chart of Average Medical Cost by Age
ages = df.Age
avg_medical_age = df.groupby('Age')['MedicalCost'].mean()
plt.title('Average Medical Cost by Age')
plt.xlabel('Age')
plt.ylabel('Average Medical Cost')
plt.plot(sorted(ages.unique()), avg_medical_age, color='green')
plt.scatter(sorted(ages.unique()), avg_medical_age, color='green')
plt.grid(True)
```

```python
# Proportion of smokers vs non-smokers by region
region = df.Region
smokerCounts = df.Smoker

#test = df.groupby('Smoker')['Region']
smokeProportions = df.groupby('Region')['Smoker'].value_counts(normalize=True).unstack()
#smokeProportions = smokeProportions[['yes', 'no']]
smokeProportions.plot(kind='bar', stacked=True, color=['lightskyblue', 'lightcoral'])
plt.title('Proportion of Smokers vs. Non-Smokers by Region')
plt.ylabel('Proportion')
```
